1. What is hypothesis testing in statistics?
        -> Hypothesis testing is a statistical method used to determine if there is enough evidence in a sample data to draw conclusions about a population. It involves formulating a null hypothesis (H0) and an alternative hypothesis (H1), collecting and analyzing data, and then making a decision to either reject or fail to reject the null hypothesis based on the evidence.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
        -> The null hypothesis (H0) is a statement of no effect, no difference, or no relationship between variables. It's the default assumption that we try to find evidence against. The alternative hypothesis (H1 or Ha) is a statement that contradicts the null hypothesis, suggesting there is an effect, difference, or relationship. It's what we conclude if we reject the null hypothesis.

3. What is the significance level in hypothesis testing, and why is it important?
        -> The significance level, often denoted by alpha (α), is the probability of rejecting the null hypothesis when it is actually true. It's the threshold for determining statistical significance. A common significance level is 0.05 (or 5%), meaning there's a 5% risk of incorrectly rejecting the null hypothesis. It's important because it sets the standard for how much evidence is needed to conclude that a result is statistically significant.

4. What does a P-value represent in hypothesis testing?
        -> The P-value, or probability value, in hypothesis testing is the probability of observing a test statistic as extreme as, or more extreme than, the one calculated from your sample data, assuming that the null hypothesis is true. It quantifies the strength of the evidence against the null hypothesis. A smaller P-value indicates stronger evidence against the null hypothesis.

6. What are Type 1 and Type 2 errors in hypothesis testing?
        -> In hypothesis testing, there are two main types of errors you can make:

        **Type 1 Error (False Positive)**: This occurs when you reject the null hypothesis (H0) when it is actually true. The probability of making a Type 1 error is equal to the significance level (alpha, α).
        **Type 2 Error (False Negative)**: This occurs when you fail to reject the null hypothesis (H0) when the alternative hypothesis (H1) is actually true. The probability of making a Type 2 error is often denoted by beta (β).

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
        -> In hypothesis testing, the choice between a one-tailed and a two-tailed test depends on the nature of the alternative hypothesis:

        **One-tailed test**: Used when the alternative hypothesis specifies a direction of the effect (e.g., the mean is greater than a certain value, or the mean is less than a certain value). The rejection region is entirely in one tail of the sampling distribution.
        **Two-tailed test**: Used when the alternative hypothesis does not specify a direction (e.g., the mean is different from a certain value). The rejection region is split between both tails of the sampling distribution.

8. What is the Z-test, and when is it used in hypothesis testing?

        It is used in hypothesis testing when:
        * The population standard deviation is known.
        * The sample size is large (generally considered to be n ≥ 30).
        * The data is approximately normally distributed.

10. What is the T-distribution, and when should it be used instead of the normal distribution?
        -> The T-distribution, also known as the Student's T-distribution, is a probability distribution that arises when estimating the mean of a normally distributed population in situations where the sample size is small and the population's standard deviation is unknown.
        It should be used instead of the normal distribution (Z-distribution) in the following scenarios:
        * When the population standard deviation is unknown: This is the most common reason to use the T-distribution. Since the population standard deviation is unknown, we have to estimate it using the sample standard deviation. This introduces more variability, and the T-distribution accounts for this increased uncertainty, especially with smaller sample sizes.

        * When the sample size is small: Although there's no strict cutoff, the T-distribution is particularly important for small sample sizes (typically n < 30). For larger sample sizes, the T-distribution closely approximates the normal distribution.



11. What is the difference between a Z-test and a T-test?
        -> The main difference between a Z-test and a T-test lies in the information available about the population standard deviation and the sample size.

        **Z-test**: Used when the population standard deviation ($\sigma$$\sigma$) is known or when the sample size is large (typically n ≥ 30), allowing the sample standard deviation to serve as a good estimate for the population standard deviation. It uses the standard normal distribution (Z-distribution) to determine critical values and p-values.
        **T-test**: Used when the population standard deviation ($\sigma$$\sigma$) is unknown and must be estimated from the sample standard deviation (s). It is particularly important when the sample size is small (typically n < 30), as the T-distribution accounts for the increased uncertainty due to estimating the population standard deviation from a small sample. The T-test uses the T-distribution, which has heavier tails than the normal distribution, especially with fewer degrees of freedom (related to sample size).

12. What is the T-test, and how is it used in hypothesis testing?
        -> The T-test is a statistical hypothesis test used to determine if there is a significant difference between the means of two groups, or between the mean of a single group and a known value (a one-sample T-test). It is used when the population standard deviation is unknown and the sample size is relatively small (typically less than 30), though it can be used with larger sample sizes as well.
        It's used in hypothesis testing:
        * Formulate Hypotheses
        * Calculate the Test Statistic
        * Determine Degrees of Freedom
        * Determine the Critical Value or P-value

13. What is the relationship between Z-test and T-test in hypothesis testing?
        -> The Z-test and T-test are both statistical hypothesis tests used to compare means, but they are related and used under different circumstances. Here's the relationship:
        * Both are used to compare means
        * The T-test is an adaptation of the Z-test for unknown population standard deviation.
        * The T-distribution accounts for the uncertainty of estimating the standard deviation.
        * As sample size increases, the T-distribution approaches the Z-distribution.

14. What is a confidence interval, and how is it used to interpret statistical results?
        -> A confidence interval is a range of values that is likely to contain a population parameter (such as the mean) with a certain level of confidence. It's calculated from sample data and provides a range of plausible values for the unknown population parameter.
        It's used to interpret statistical results:
        * Estimating the Population Parameter.
        * Quantifying Uncertainty.
        * Hypothesis Testing Connection.
        * Communicating Results.

15. What is the margin of error, and how does it affect the confidence interval?
        -> The margin of error is a crucial component of a confidence interval. It is the "plus or minus" amount that you add and subtract from the point estimate (like the sample mean) to create the interval.
        Mathematically, the margin of error is typically calculated as:
        $$ \text{Margin of Error} = \text{Critical Value} \times \text{Standard Error} $$$$ \text{Margin of Error} = \text{Critical Value} \times \text{Standard Error} $$

        The margin of error directly determines the width of the confidence interval.
        It affect the confidence interval by:
        Confidence Interval = Point Estimate ± Margin of Error
        So, the confidence interval spans from (Point Estimate - Margin of Error) to (Point Estimate + Margin of Error).


16. How is Bayes' Theorem used in statistics, and what is its significance?
        -> Bayes' Theorem is a fundamental concept in probability theory that describes how to update the probability of a hypothesis based on new evidence. In statistics, it is the cornerstone of Bayesian inference.
        The theorem is stated as:
        $$ P(A|B) = \frac{P(B|A) \times P(A)}{P(B)} $$$$ P(A|B) = \frac{P(B|A) \times P(A)}{P(B)} $$

        Bayes' Theorem allows us to start with an initial belief about a parameter or hypothesis (the prior probability) and update that belief as we collect data (the evidence). The result is a posterior probability distribution, which represents our updated belief about the parameter or hypothesis after considering the data.

        Significance of Bayes' Theorem in Statistics:
        * Foundation of Bayesian Inference
        * Natural for Sequential Learning.
        * Provides a Probability Distribution for Parameters.
        * Applications: It is widely used in various fields, including machine learning, medical diagnosis, finance, and scientific research, especially for complex modeling and situations with limited data.

17. What is the Chi-square distribution, and when is it used?
        -> The Chi-square ($\chi^2$$\chi^2$) distribution is a continuous probability distribution that arises in statistics, particularly in hypothesis testing and in constructing confidence intervals. It is the distribution of a sum of the squares of independent standard normal random variables.
        The Chi-square distribution is primarily used in hypothesis testing for:
        * Chi-square Goodness-of-Fit Test

18. What is the Chi-square goodness of fit test, and how is it applied?
        -> The Chi-square goodness-of-fit test is a statistical hypothesis test used to determine if an observed frequency distribution of a single categorical variable differs significantly from an expected frequency distribution. In simpler terms, it checks if the observed data "fits" a theoretical or hypothesized distribution.
        It is applied:
        **Formulate Hypotheses**:
        Null Hypothesis (H0): The observed frequency distribution matches the expected frequency distribution. There is no significant difference.
        Alternative Hypothesis (H1): The observed frequency distribution does not match the expected frequency distribution. There is a significant difference.
        **Collect Observed Data**: Gather the observed frequencies for each category of the categorical variable.
        **Determine Degrees of Freedom** : The degrees of freedom for the goodness-of-fit test are calculated as the number of categories minus the number of parameters estimated from the data minus 1

19. What is the F-distribution, and when is it used in hypothesis testing?
        -> The F-distribution is a continuous probability distribution that arises frequently in the context of comparing variances, particularly in Analysis of Variance (ANOVA) and regression analysis. It is named after Sir Ronald Fisher.
        The F-distribution is primarily used as the sampling distribution for the F-statistic, which is a test statistic used to compare variances or assess the overall significance of a statistical model. Its main applications in hypothesis testing include:
        **Analysis of Variance (ANOVA)**: This is the most common use. ANOVA tests whether there are any statistically significant differences between the means of three or more independent groups. The F-statistic in ANOVA compares the variance between the groups to the variance within the groups. If the variance between groups is significantly larger than the variance within groups, it suggests that at least one group mean is different from the others.
        **Regression Analysis**: In regression, the F-test is used to assess the overall significance of the regression model. It tests whether the model as a whole explains a significant amount of the variation in the dependent variable, compared to a model with no independent variables.


20. What is an ANOVA test, and what are its assumptions?
        -> An ANOVA (Analysis of Variance) test is a statistical hypothesis test used to determine if there are any statistically significant differences between the means of three or more independent groups. It's an extension of the t-test, which is used to compare the means of only two groups.
        ANOVA is based on several key assumptions. If these assumptions are violated, the results of the ANOVA test may not be reliable. The main assumptions are:
        **Independence of Observations**: The observations within and between groups must be independent. This means that the value of one observation does not influence the value of another observation. This is often ensured through proper experimental design and random sampling.
        **Normality**: The data within each group should be approximately normally distributed. ANOVA is relatively robust to minor violations of this assumption, especially with larger sample sizes due to the Central Limit Theorem. You can check this assumption using histograms, Q-Q plots, or normality tests.

21. What are the different types of ANOVA tests?
        -> There are several different types of ANOVA tests, depending on the number of independent variables (factors) and the design of the study (e.g., independent groups vs. repeated measures). Here are some of the most common types:
        **One-Way ANOVA**: Used when you have one categorical independent variable (factor) and one continuous dependent variable. It tests whether there are statistically significant differences between the means of three or more independent groups defined by the levels of the independent variable.
        **Two-Way ANOVA**: Used when you have two categorical independent variables (factors) and one continuous dependent variable. It allows you to examine the main effect of each factor individually, as well as the interaction effect between the two factors.
        **Repeated Measures ANOVA**: Used when you have one or more categorical independent variables and one continuous dependent variable, but the same subjects are measured under multiple conditions or at multiple time points. This means the observations within each subject are not independent.
        **Mixed ANOVA** (or Split-Plot ANOVA): Used when you have a combination of between-subjects factors (different subjects in each group) and within-subjects factors

22. What is the F-test, and how does it relate to hypothesis testing?
        -> The F-test is any statistical test that uses the F-distribution to calculate a P-value and determine whether to reject the null hypothesis. It's most commonly used to compare variances or to assess the overall significance of a statistical model.
        In hypothesis testing, the F-test is used to evaluate the ratio of two variances. This ratio follows the F-distribution under the null hypothesis. The F-statistic, calculated from your data, is compared to critical values from the F-distribution or used to calculate a P-value.

```python
# Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint


low = 1
high = 7

dist = randint(low, high)

x = np.arange(low, high)

pmf_values = dist.pmf(x)

plt.stem(x, pmf_values)
plt.title('Discrete Uniform Distribution PMF')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.xticks(x)
plt.ylim(0, 1)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
# Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

def bernoulli_pdf(k, p):
  """
  Calculates the probability distribution function (PDF) of a Bernoulli distribution.

  Args:
    k: The outcome (0 or 1).
    p: The probability of success (between 0 and 1).

  Returns:
    The probability of the given outcome k.
  """
  if k == 1:
    return p
  elif k == 0:
    return 1 - p
  else:
    return 0

p = 0.7
print(f"PDF at k=1 for p={p}: {bernoulli_pdf(1, p)}")
print(f"PDF at k=0 for p={p}: {bernoulli_pdf(0, p)}")
print(f"PDF at k=2 for p={p}: {bernoulli_pdf(2, p)}")
```

```python
# Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

import numpy as np
import matplotlib.pyplot as plt

lambda_param = 0.5
beta = 1/lambda_param
population_mean = beta
population_std = beta

num_samples = 1000

sample_size = 30

sample_means = []

for _ in range(num_samples):
    sample = np.random.exponential(scale=beta, size=sample_size)
    sample_mean = np.mean(sample)
    sample_means.append(sample_mean)

sample_means = np.array(sample_means)

plt.figure(figsize=(10, 6))
plt.hist(sample_means, bins=30, density=True, edgecolor='black', alpha=0.7, label='Distribution of Sample Means')


from scipy.stats import norm
x = np.linspace(sample_means.min(), sample_means.max(), 100)
y = norm.pdf(x, loc=population_mean, scale=population_std / np.sqrt(sample_size))
plt.plot(x, y, color='red', linestyle='--', label='Theoretical Normal Distribution (CLT)')

plt.title('Demonstration of the Central Limit Theorem')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency / Probability Density')
plt.legend()
plt.grid(axis='y', alpha=0.75)
plt.show()

print(f"Population Mean (Theoretical): {population_mean:.4f}")
print(f"Mean of Sample Means (Simulated): {np.mean(sample_means):.4f}")
print(f"Standard Error (Theoretical): {population_std / np.sqrt(sample_size):.4f}")
print(f"Standard Deviation of Sample Means (Simulated): {np.std(sample_means):.4f}")
```

```python
# Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.

import numpy as np
from scipy.stats import norm

def calculate_and_compare_zscore(data_point, dataset):
  """
  Calculates the Z-score for a data point and compares it to the standard normal distribution.

  Args:
    data_point: The individual data point.
    dataset: The dataset (NumPy array or list) from which the data point comes.
  """
  mean = np.mean(dataset)
  std_dev = np.std(dataset)

  if std_dev == 0:
      print("Cannot calculate Z-score: Standard deviation is zero.")
      return None

  z_score = (data_point - mean) / std_dev

  print(f"Data Point: {data_point}")
  print(f"Dataset Mean: {mean:.2f}")
  print(f"Dataset Standard Deviation: {std_dev:.2f}")
  print(f"Calculated Z-score: {z_score:.4f}")


  prob_less_than_z = norm.cdf(z_score)
  prob_greater_than_z = 1 - prob_less_than_z

  print(f"Probability of a standard normal variable being less than this Z-score: {prob_less_than_z:.4f}")
  print(f"Probability of a standard normal variable being greater than this Z-score: {prob_greater_than_z:.4f}")

dataset1 = np.array([10, 12, 15, 11, 13, 14, 16, 12, 10, 15])
data_point1 = 16
calculate_and_compare_zscore(data_point1, dataset1)

print("-" * 30)

dataset2 = np.random.normal(loc=50, scale=10, size=100)
data_point2 = 65
calculate_and_compare_zscore(data_point2, dataset2)
```

```python
# Implement hypothesis testing using Z-statistics for a sample dataset.

import numpy as np
from scipy import stats

sample_data = np.array([62, 65, 63, 66, 61, 67, 64, 68, 60, 69])
population_mean_h0 = 60
population_std_dev = 5
alpha = 0.05

sample_mean = np.mean(sample_data)
sample_size = len(sample_data)

z_statistic = (sample_mean - population_mean_h0) / (population_std_dev / np.sqrt(sample_size))

p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Hypothesized Population Mean (H0): {population_mean_h0}")
print(f"Population Standard Deviation: {population_std_dev}")
print(f"Z-statistic: {z_statistic:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Significance Level (alpha): {alpha}")

if p_value < alpha:
    print("Result: Reject the null hypothesis.")
    print("Conclusion: There is sufficient evidence to conclude that the sample mean is significantly different from the hypothesized population mean.")
else:
    print("Result: Fail to reject the null hypothesis.")
    print("Conclusion: There is not enough evidence to conclude that the sample mean is significantly different from the hypothesized population mean.")
```

```python
# Create a confidence interval for a dataset using Python and interpret the result.

import numpy as np
from scipy import stats

data = np.array([12, 15, 11, 13, 14, 16, 12, 10, 15, 13, 14, 12])

confidence_level = 0.95
alpha = 1 - confidence_level

sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
sample_size = len(data)

standard_error = sample_std / np.sqrt(sample_size)


degrees_freedom = sample_size - 1
t_critical = stats.t.ppf(1 - alpha / 2, degrees_freedom)
margin_of_error = t_critical * standard_error

confidence_interval_lower = sample_mean - margin_of_error
confidence_interval_upper = sample_mean + margin_of_error

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Standard Deviation: {sample_std:.2f}")
print(f"Sample Size: {sample_size}")
print(f"Confidence Level: {confidence_level}")
print(f"Degrees of Freedom: {degrees_freedom}")
print(f"T-critical value: {t_critical:.4f}")
print(f"Margin of Error: {margin_of_error:.4f}")
print(f"Confidence Interval: ({confidence_interval_lower:.4f}, {confidence_interval_upper:.4f})")

print("\nInterpretation:")
print(f"We are {confidence_level * 100}% confident that the true population mean lies within the interval ({confidence_interval_lower:.4f}, {confidence_interval_upper:.4f}).")
```

```python
# Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_and_visualize_errors(population_mean_h0, population_mean_h1, population_std, sample_size, alpha):
    """
    Calculates and visualizes Type 1 and Type 2 errors for a one-sample Z-test.

    Args:
        population_mean_h0: Population mean under the null hypothesis.
        population_mean_h1: Population mean under the alternative hypothesis.
        population_std: Population standard deviation.
        sample_size: The size of the sample.
        alpha: The significance level (Type 1 error rate).
    """
    standard_error = population_std / np.sqrt(sample_size)

    z_critical = norm.ppf(1 - alpha / 2)

    critical_sample_mean_lower = population_mean_h0 - z_critical * standard_error
    critical_sample_mean_upper = population_mean_h0 + z_critical * standard_error

    type1_error = alpha


    z_lower_under_h1 = (critical_sample_mean_lower - population_mean_h1) / standard_error
    z_upper_under_h1 = (critical_sample_mean_upper - population_mean_h1) / standard_error


    type2_error = norm.cdf(z_upper_under_h1) - norm.cdf(z_lower_under_h1)

    power = 1 - type2_error

    print(f"Population Mean (H0): {population_mean_h0}")
    print(f"Population Mean (H1): {population_mean_h1}")
    print(f"Population Standard Deviation: {population_std}")
    print(f"Sample Size: {sample_size}")
    print(f"Significance Level (alpha): {alpha}")
    print(f"Critical Z-value: ±{z_critical:.4f}")
    print(f"Critical Sample Mean Range (Acceptance Region for H0): ({critical_sample_mean_lower:.4f}, {critical_sample_mean_upper:.4f})")
    print(f"Type 1 Error (alpha): {type1_error:.4f}")
    print(f"Type 2 Error (beta): {type2_error:.4f}")
    print(f"Power (1 - beta): {power:.4f}")

    plt.figure(figsize=(12, 6))
    x = np.linspace(min(population_mean_h0, population_mean_h1) - 4 * standard_error,
                    max(population_mean_h0, population_mean_h1) + 4 * standard_error, 200)

    plt.plot(x, norm.pdf(x, loc=population_mean_h0, scale=standard_error), label='Distribution under H0', color='blue')
    x_reject_h0_left = x[x <= critical_sample_mean_lower]
    plt.fill_between(x_reject_h0_left, norm.pdf(x_reject_h0_left, loc=population_mean_h0, scale=standard_error),
                     color='red', alpha=0.3, label=f'Type 1 Error (α = {type1_error:.4f})')
    x_reject_h0_right = x[x >= critical_sample_mean_upper]
    plt.fill_between(x_reject_h0_right, norm.pdf(x_reject_h0_right, loc=population_mean_h0, scale=standard_error),
                     color='red', alpha=0.3)

    plt.plot(x, norm.pdf(x, loc=population_mean_h1, scale=standard_error), label='Distribution under H1', color='green')
    x_type2_error = x[(x >= critical_sample_mean_lower) & (x <= critical_sample_mean_upper)]
    plt.fill_between(x_type2_error, norm.pdf(x_type2_error, loc=population_mean_h1, scale=standard_error),
                     color='orange', alpha=0.5, label=f'Type 2 Error (β = {type2_error:.4f})')

    x_power_left = x[x <= critical_sample_mean_lower]
    plt.fill_between(x_power_left, norm.pdf(x_power_left, loc=population_mean_h1, scale=standard_error),
                     color='purple', alpha=0.3, label=f'Power (1-β = {power:.4f})')
    x_power_right = x[x >= critical_sample_mean_upper]
    plt.fill_between(x_power_right, norm.pdf(x_power_right, loc=population_mean_h1, scale=standard_error),
                     color='purple', alpha=0.3)


    plt.axvline(critical_sample_mean_lower, color='black', linestyle='--', label='Critical Value (Sample Mean)')
    plt.axvline(critical_sample_mean_upper, color='black', linestyle='--')

    plt.title('Visualization of Type 1 and Type 2 Errors')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

calculate_and_visualize_errors(population_mean_h0=50, population_mean_h1=52, population_std=10, sample_size=30, alpha=0.05)

print("\n" + "="*30 + "\n")

calculate_and_visualize_errors(population_mean_h0=100, population_mean_h1=101, population_std=15, sample_size=50, alpha=0.01)
```

```python
# Write a Python program to perform an independent T-test and interpret the results.

import numpy as np
from scipy import stats

group1_data = np.array([85, 88, 90, 82, 87, 86, 89, 91, 83, 88])
group2_data = np.array([78, 80, 82, 75, 79, 81, 83, 77, 76, 80])

ttest_independent_result = stats.ttest_ind(group1_data, group2_data)

t_statistic = ttest_independent_result.statistic
p_value = ttest_independent_result.pvalue

alpha = 0.05

print(f"Group 1 Mean: {np.mean(group1_data):.2f}")
print(f"Group 2 Mean: {np.mean(group2_data):.2f}")
print(f"T-statistic: {t_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print(f"Result: Reject the null hypothesis at α = {alpha}. There is a significant difference between the means of the two groups.")
else:
    print(f"Result: Fail to reject the null hypothesis at α = {alpha}. There is no significant difference between the means of the two groups.")
```

```python
# Perform a paired sample T-test using Python and visualize the comparison results.

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

np.random.seed(42)
before_treatment = np.random.normal(loc=70, scale=5, size=20)

after_treatment = before_treatment + np.random.normal(loc=2, scale=3, size=20)

ttest_paired_result = stats.ttest_rel(before_treatment, after_treatment)

t_statistic = ttest_paired_result.statistic
p_value = ttest_paired_result.pvalue

alpha = 0.05

print(f"Mean 'Before Treatment': {np.mean(before_treatment):.2f}")
print(f"Mean 'After Treatment': {np.mean(after_treatment):.2f}")
print(f"T-statistic (paired): {t_statistic:.4f}")
print(f"P-value (paired): {p_value:.4f}")

if p_value < alpha:
    print(f"Result: Reject the null hypothesis at α = {alpha}. There is a significant difference between the 'Before' and 'After' treatment means.")
else:
    print(f"Result: Fail to reject the null hypothesis at α = {alpha}. There is no significant difference between the 'Before' and 'After' treatment means.")

plt.figure(figsize=(8, 6))
plt.boxplot([before_treatment, after_treatment], labels=['Before Treatment', 'After Treatment'])
plt.title('Comparison of Before and After Treatment')
plt.ylabel('Measurement Value')
plt.grid(axis='y', alpha=0.75)
plt.show()

differences = after_treatment - before_treatment
plt.figure(figsize=(8, 6))
plt.hist(differences, bins=10, edgecolor='black', alpha=0.7)
plt.title('Distribution of Differences (After - Before)')
plt.xlabel('Difference in Value')
plt.ylabel('Frequency')
plt.axvline(0, color='red', linestyle='--', label='Zero Difference')
plt.legend()
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
# Simulate data and perform both Z-test and T-test, then compare the results using Python.

import numpy as np
from scipy import stats

np.random.seed(42)
population_mean = 100
population_std = 10
sample_size = 30

sample_data = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)

hypothesized_mean = 103


z_statistic = (np.mean(sample_data) - hypothesized_mean) / (population_std / np.sqrt(sample_size))
p_value_z = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

print("--- Z-test Results ---")
print(f"Sample Mean: {np.mean(sample_data):.2f}")
print(f"Hypothesized Mean (H0): {hypothesized_mean}")
print(f"Population Standard Deviation: {population_std}")
print(f"Z-statistic: {z_statistic:.4f}")
print(f"P-value (Z-test): {p_value_z:.4f}")
alpha = 0.05
if p_value_z < alpha:
    print("Z-test Result: Reject H0 (significant difference).")
else:
    print("Z-test Result: Fail to reject H0 (no significant difference).")

print("\n" + "="*30 + "\n")


sample_std = np.std(sample_data, ddof=1)

t_statistic = (np.mean(sample_data) - hypothesized_mean) / (sample_std / np.sqrt(sample_size))
degrees_freedom = sample_size - 1
p_value_t = 2 * (1 - stats.t.cdf(abs(t_statistic), df=degrees_freedom))

print("--- T-test Results ---")
print(f"Sample Mean: {np.mean(sample_data):.2f}")
print(f"Hypothesized Mean (H0): {hypothesized_mean}")
print(f"Sample Standard Deviation: {sample_std:.2f}")
print(f"T-statistic: {t_statistic:.4f}")
print(f"Degrees of Freedom: {degrees_freedom}")
print(f"P-value (T-test): {p_value_t:.4f}")
if p_value_t < alpha:
    print("T-test Result: Reject H0 (significant difference).")
else:
    print("T-test Result: Fail to reject H0 (no significant difference).")

print("\n" + "="*30 + "\n")

print("--- Comparison ---")
print(f"Z-statistic: {z_statistic:.4f}, P-value (Z-test): {p_value_z:.4f}")
print(f"T-statistic: {t_statistic:.4f}, P-value (T-test): {p_value_t:.4f}")

print("\nObservations:")
if sample_size >= 30 and abs(z_statistic - t_statistic) < 0.5: # Arbitrary threshold for "close"
    print("For a sufficiently large sample size (n >= 30), the Z-test and T-test results are generally close when the population standard deviation is known or the sample standard deviation is a good estimate.")
    print("The T-test is more appropriate when the population standard deviation is unknown, as it accounts for the additional uncertainty.")
elif sample_size < 30:
     print("For a small sample size (n < 30), the T-test is generally preferred when the population standard deviation is unknown.")
     print("The T-distribution has heavier tails than the normal distribution, which is more appropriate for smaller samples and accounts for the increased variability in the sample standard deviation.")
else:
     print("The Z-test assumes the population standard deviation is known, while the T-test uses the sample standard deviation. The results may differ, especially with smaller sample sizes or when the sample standard deviation is not a good estimate of the population standard deviation.")
```

```python
# Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

import numpy as np
from scipy import stats

def calculate_confidence_interval(data, confidence_level=0.95):
  """
  Calculates the confidence interval for the mean of a dataset.

  Args:
    data: A NumPy array or list of numerical data.
    confidence_level: The desired confidence level (e.g., 0.95 for 95%).

  Returns:
    A tuple containing the lower and upper bounds of the confidence interval.
  """
  sample_mean = np.mean(data)
  sample_std = np.std(data, ddof=1)
  sample_size = len(data)

  if sample_size < 2:
    print("Error: At least two data points are required to calculate a confidence interval.")
    return None, None

  standard_error = sample_std / np.sqrt(sample_size)

  degrees_freedom = sample_size - 1
  alpha = 1 - confidence_level
  t_critical = stats.t.ppf(1 - alpha / 2, degrees_freedom)

  margin_of_error = t_critical * standard_error

  confidence_interval_lower = sample_mean - margin_of_error
  confidence_interval_upper = sample_mean + margin_of_error

  return confidence_interval_lower, confidence_interval_upper

data = np.array([12, 15, 11, 13, 14, 16, 12, 10, 15, 13, 14, 12])
lower_bound, upper_bound = calculate_confidence_interval(data)

if lower_bound is not None and upper_bound is not None:
  print(f"Sample Data: {data}")
  print(f"Confidence Level: 95%")
  print(f"Confidence Interval: ({lower_bound:.4f}, {upper_bound:.4f})")

  print("\nSignificance of the Confidence Interval:")
  print(f"A {95}% confidence interval means that if we were to take many random samples from the same population and calculate a confidence interval for each sample, approximately 95% of those intervals would contain the true population mean.")
  print(f"Based on this specific sample, we are {95}% confident that the true population mean lies within the range of {lower_bound:.4f} to {upper_bound:.4f}.")
  print("It provides a range of plausible values for the population mean, rather than a single point estimate.")

print("-" * 30)

data2 = np.random.normal(loc=70, scale=10, size=50)
lower_bound2, upper_bound2 = calculate_confidence_interval(data2, confidence_level=0.90)

if lower_bound2 is not None and upper_bound2 is not None:
  print(f"Sample Data (first 10 elements): {data2[:10]}")
  print(f"Confidence Level: 90%")
  print(f"Confidence Interval: ({lower_bound2:.4f}, {upper_bound2:.4f})")

  print("\nSignificance of the Confidence Interval:")
  print(f"We are {90}% confident that the true population mean lies within the range of {lower_bound2:.4f} to {upper_bound2:.4f}.")
```

```python
# Write a Python program to calculate the margin of error for a given confidence level using sample data.

import numpy as np
from scipy import stats

def calculate_margin_of_error(data, confidence_level=0.95):
  """
  Calculates the margin of error for the mean of a dataset.

  Args:
    data: A NumPy array or list of numerical data.
    confidence_level: The desired confidence level (e.g., 0.95 for 95%).

  Returns:
    The margin of error.
  """
  sample_mean = np.mean(data)
  sample_std = np.std(data, ddof=1)
  sample_size = len(data)

  if sample_size < 2:
    print("Error: At least two data points are required to calculate margin of error.")
    return None

  standard_error = sample_std / np.sqrt(sample_size)

  degrees_freedom = sample_size - 1
  alpha = 1 - confidence_level
  t_critical = stats.t.ppf(1 - alpha / 2, degrees_freedom)

  margin_of_error = t_critical * standard_error

  return margin_of_error

data = np.array([12, 15, 11, 13, 14, 16, 12, 10, 15, 13, 14, 12])
margin_error = calculate_margin_of_error(data)

if margin_error is not None:
  print(f"Sample Data: {data}")
  print(f"Confidence Level: 95%")
  print(f"Calculated Margin of Error: {margin_error:.4f}")

print("-" * 30)

data2 = np.random.normal(loc=70, scale=10, size=50)
margin_error2 = calculate_margin_of_error(data2, confidence_level=0.90)

if margin_error2 is not None:
  print(f"Sample Data (first 10 elements): {data2[:10]}")
  print(f"Confidence Level: 90%")
  print(f"Calculated Margin of Error: {margin_error2:.4f}")
```

```python
# Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

prior_disease = 0.001


likelihood_positive_disease = 0.99


likelihood_positive_no_disease = 0.05

prob_no_disease = 1 - prior_disease


prob_positive = (likelihood_positive_disease * prior_disease) + (likelihood_positive_no_disease * prob_no_disease)


posterior_disease_positive = (likelihood_positive_disease * prior_disease) / prob_positive

print(f"Prior Probability of Disease (P(Disease)): {prior_disease:.4f}")
print(f"Likelihood of Positive Test given Disease (P(Positive | Disease)): {likelihood_positive_disease:.4f}")
print(f"Likelihood of Positive Test given No Disease (P(Positive | No Disease)): {likelihood_positive_no_disease:.4f}")
print(f"Probability of Positive Test (P(Positive)): {prob_positive:.4f}")
print(f"Posterior Probability of Disease given Positive Test (P(Disease | Positive)): {posterior_disease_positive:.4f}")


print("\n--- Explanation of Bayesian Inference ---")
print("Bayesian inference is a method of statistical inference in which Bayes' theorem is used to update the probability for a hypothesis as more evidence or information becomes available.")
print("In this example, we started with a 'Prior Probability' of having the disease (P(Disease)) based on general population data.")
print("The medical test provides new 'Evidence' (a positive test result).")
print("We use the 'Likelihood' (P(Positive | Disease) and P(Positive | No Disease)) which tells us how likely the evidence is under different hypotheses (having or not having the disease).")
print("Bayes' Theorem combines the Prior Probability and the Likelihood to calculate the 'Posterior Probability' (P(Disease | Positive)).")
print("The Posterior Probability is the updated probability of having the disease given the new evidence (the positive test result).")
print("As you can see from the results, even with a positive test from a seemingly good test (high sensitivity, decent specificity), the probability of actually having a rare disease is still relatively low due to the low prior probability.")
```

```python
# Perform a Chi-square test for independence between two categorical variables in Python.

import numpy as np
from scipy.stats import chi2_contingency


observed_data = np.array([[20, 15, 25],
                          [30, 40, 35]])

chi2_statistic, p_value, dof, expected_frequencies = chi2_contingency(observed_data)

print("Observed Data (Contingency Table):")
print(observed_data)
print("\nChi-square Statistic:", chi2_statistic)
print("P-value:", p_value)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:")
print(expected_frequencies)

alpha = 0.05
print(f"\nSignificance Level (alpha): {alpha}")

if p_value < alpha:
    print("Result: Reject the null hypothesis.")
    print("Conclusion: There is a statistically significant association between the two categorical variables.")
else:
    print("Result: Fail to reject the null hypothesis.")
    print("Conclusion: There is no statistically significant association between the two categorical variables.")
```

```python
# Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

import numpy as np
from scipy.stats import chisquare


observed_frequencies = np.array([10, 12, 8, 15, 9, 6])

total_rolls = np.sum(observed_frequencies)
num_categories = len(observed_frequencies)
expected_frequencies = np.full(num_categories, total_rolls / num_categories)

print("Observed Frequencies:", observed_frequencies)
print("Expected Frequencies (for a fair die):", expected_frequencies)

chi2_statistic, p_value = chisquare(observed_frequencies, f_exp=expected_frequencies)

print("\nChi-square Statistic:", chi2_statistic)
print("P-value:", p_value)

alpha = 0.05
print(f"\nSignificance Level (alpha): {alpha}")

if p_value < alpha:
    print("Result: Reject the null hypothesis.")
    print("Conclusion: The observed distribution is significantly different from the expected distribution (the die is likely not fair).")
else:
    print("Result: Fail to reject the null hypothesis.")
    print("Conclusion: The observed distribution is not significantly different from the expected distribution (the die is likely fair).")
```
