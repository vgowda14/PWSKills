# What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.
        -> Anomaly detection is the process of identifying data points, events, or observations that deviate significantly from the norm or expected behavior. These deviations, or anomalies, can indicate unusual occurrences such as fraud, equipment malfunctions, or rare events.

There are three main types of anomalies:

1.  **Point Anomalies:** These are single data points that are significantly different from the rest of the data.
    *   **Example:** A sudden, unusually high credit card transaction in a user's account history, where all previous transactions were small.

2.  **Contextual Anomalies:** These are data points that are anomalous only within a specific context. The same data point might not be considered an anomaly in a different context.
    *   **Example:** A temperature reading of 30°C in a desert is normal, but the same reading in the Arctic would be a contextual anomaly.

3.  **Collective Anomalies:** These are a set of data points that, when considered together, are anomalous, even though individual data points within the set may not be anomalous.
    *   **Example:** A series of small, seemingly insignificant network requests that, when analyzed collectively, form a pattern indicative of a denial-of-service attack.

# Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.
        -> Isolation Forest, DBSCAN, and Local Outlier Factor based on their approach and suitable use cases:

**1. Isolation Forest**

*   **Approach:** Isolation Forest is an ensemble tree-based method. It works by randomly selecting a feature and then randomly selecting a split value between the minimum and maximum values of the selected feature. This process recursively partitions the data. Anomalies are data points that are isolated in fewer splits, meaning they have shorter path lengths in the trees compared to normal instances.
*   **Suitable Use Cases:**
    *   High-dimensional datasets.
    *   Datasets with a large number of instances.
    *   Efficient for detecting point anomalies.
    *   Can handle irrelevant features.

**2. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)**

*   **Approach:** DBSCAN is a density-based clustering algorithm. It groups together points that are closely packed together (having many neighbors within a certain distance) and marks points that lie alone in low-density regions as outliers (anomalies). It requires two parameters: `epsilon` (the maximum distance between two samples for one to be considered as in the neighborhood of the other) and `min_samples` (the number of samples in a neighborhood for a point to be considered as a core point).
*   **Suitable Use Cases:**
    *   Discovering clusters of arbitrary shape.
    *   Identifying point anomalies (noise points).
    *   Requires careful tuning of `epsilon` and `min_samples`, which can be challenging.
    *   Sensitive to varying densities in the data.

**3. Local Outlier Factor (LOF)**

*   **Approach:** LOF is a density-based technique that measures the local deviation of density of a given data point with respect to its neighbors. It calculates the local reachability density of a point and compares it to the local reachability densities of its neighbors. A point is considered an outlier if its local density is significantly lower than that of its neighbors.
*   **Suitable Use Cases:**
    *   Detecting outliers in datasets where the density of the data varies.
    *   Identifying point anomalies.
    *   Can be computationally expensive for very large datasets.
    *   The definition of "local" is determined by the number of neighbors considered (`k`).

# What are the key components of a Time Series? Explain each with one example.
        -> The key components of a time series are:

1.  **Trend:** This is the long-term movement or direction of the data. It represents the underlying pattern of increase or decrease over time.
    *   **Example:** The gradual increase in the average global temperature over the past century.

2.  **Seasonality:** This refers to repeating patterns or cycles that occur within a fixed period, such as a day, week, month, or year. These patterns are often related to calendar events.
    *   **Example:** The increase in retail sales during the holiday season each year.

3.  **Cyclicality:** These are longer-term cycles in the data that are not of a fixed frequency. They can be influenced by economic conditions, business cycles, or other factors.
    *   **Example:** The fluctuations in the stock market over several years, which may not follow a strict annual pattern.

4.  **Irregularity (or Noise):** This represents the random, unpredictable fluctuations in the data that cannot be explained by trend, seasonality, or cyclicality.
    *   **Example:** A sudden, unexpected dip in sales due to a natural disaster.

# Define Stationary in time series. How can you test and transform a non-stationary series into a stationary one?
        -> A **stationary** time series is one whose statistical properties (like mean, variance, and autocorrelation) do not change over time. This means that the series has no trend, no seasonality, and the variance is constant over time. Stationary series are easier to model and forecast because their behavior is consistent.

**How to test for stationarity:**

There are several ways to test for stationarity:

1.  **Visual Inspection:** Plotting the time series and visually inspecting it for trends, seasonality, or changing variance. This is a good first step but not a rigorous test.
2.  **Summary Statistics:** Comparing the mean and variance of different segments of the time series. If they are significantly different, the series is likely non-stationary.
3.  **Statistical Tests:** More formal statistical tests can be used:
    *   **Augmented Dickey-Fuller (ADF) test:** This is a common test that checks for the presence of a unit root, which indicates non-stationarity. The null hypothesis is that the time series is non-stationary.
    *   **Kwiatkowski-Phillips-Schmidt-Shin (KPSS) test:** This test is also used to test for stationarity. The null hypothesis is that the time series is stationary.

**How to transform a non-stationary series into a stationary one:**

If a time series is non-stationary, you can often transform it to make it stationary. Common transformations include:

1.  **Differencing:** This involves subtracting the previous observation from the current observation. This can remove trends and seasonality. First-order differencing (subtracting the value at t-1 from the value at t) is common for removing trends. Seasonal differencing (subtracting the value at t-k, where k is the seasonal period) can remove seasonality.
2.  **Log Transformation:** Taking the logarithm of the series can help stabilize the variance. This is useful when the variance increases with the mean.
3.  **Seasonal Decomposition:** Decomposing the series into its trend, seasonal, and residual components can help identify and remove the non-stationary parts. You can then model the stationary residual component.
4.  **Detrending and Deseasonalizing:** Explicitly removing the estimated trend and seasonal components from the series.

The choice of transformation depends on the nature of the non-stationarity observed in the series. Often, a combination of these techniques is needed.

#  Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of structure and application.
        -> **1. AR (Autoregressive) Model**

*   **Structure:** An AR model predicts future values based on a linear combination of past values. It assumes that the current value of the time series is dependent on its own previous values. An AR(p) model uses the past 'p' observations.
*   **Application:** Suitable for time series where the current value is strongly correlated with past values, exhibiting momentum or persistence. Examples include predicting stock prices or temperature based on recent history.

**2. MA (Moving Average) Model**

*   **Structure:** An MA model predicts future values based on a linear combination of past forecast errors (or white noise terms). It models the dependency between an observation and a residual error from a moving average model applied to lagged observations. An MA(q) model uses the past 'q' forecast errors.
*   **Application:** Suitable for time series where the current value is influenced by past shocks or random disturbances. Examples include modeling the impact of random events on sales figures or other metrics.

**3. ARIMA (Autoregressive Integrated Moving Average) Model**

*   **Structure:** ARIMA combines the AR and MA models with an "Integrated" component. The "Integrated" part (I) involves differencing the time series to make it stationary. An ARIMA(p, d, q) model includes 'p' autoregressive terms, 'd' degrees of differencing, and 'q' moving average terms.
*   **Application:** A versatile model for non-stationary time series that can capture trends (through differencing), seasonality (in some cases, although SARIMA is better), and short-term dependencies. Widely used for forecasting in various fields like economics, finance, and demand forecasting.

**4. SARIMA (Seasonal Autoregressive Integrated Moving Average) Model**

*   **Structure:** SARIMA extends ARIMA to handle seasonality explicitly. It includes seasonal AR, I, and MA terms in addition to the non-seasonal terms. A SARIMA(p, d, q)(P, D, Q)m model includes non-seasonal (p, d, q) components and seasonal (P, D, Q) components with a seasonal period 'm'.
*   **Application:** Specifically designed for time series with clear seasonal patterns. Excellent for forecasting data that exhibits regular cycles, such as monthly sales, quarterly reports, or daily temperature readings.

**5. SARIMAX (Seasonal Autoregressive Integrated Moving Average with Exogenous Regressors) Model**

*   **Structure:** SARIMAX is an extension of SARIMA that includes the ability to incorporate exogenous variables (external factors) into the model. These external variables can influence the time series but are not part of the series itself.
*   **Application:** Useful when external factors are known to impact the time series. For example, forecasting retail sales might include variables like marketing spend or competitor promotions as exogenous regressors. This allows the model to account for the influence of these external factors.

```python
# Load a time series dataset (e.g., AirPassengers), plot the original series, and decompose it into trend, seasonality, and residual components

import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
df = pd.read_csv(url, parse_dates=['Month'], index_col='Month')

plt.figure(figsize=(12,6))
plt.plot(df['Passengers'], color="blue")
plt.title("Monthly Air Passengers (1949–1960)")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.show()

decomposition = seasonal_decompose(df['Passengers'], model='multiplicative')

fig = decomposition.plot()
fig.set_size_inches(12, 8)
plt.show()
```

```python
# Train a SARIMA model on the monthly airline passengers dataset. Forecast the next 12 months and visualize the results.



import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX




url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
df = pd.read_csv(url, parse_dates=['Month'], index_col='Month')
df.index.freq = 'MS'

from statsmodels.tsa.stattools import adfuller
print("ADF test on original series:", adfuller(df['Passengers'])[1], "(p-value)")
print("ADF test on log-transformed series:", adfuller(np.log(df['Passengers']))[1], "(p-value)")

ts_log = np.log(df['Passengers'])


order = (1, 1, 1)
seasonal_order = (1, 1, 1, 12)

model = SARIMAX(ts_log,
                order=order,
                seasonal_order=seasonal_order,
                enforce_stationarity=False,
                enforce_invertibility=False)

res = model.fit(disp=False)
print(res.summary())

steps = 12
pred = res.get_forecast(steps=steps)
pred_mean_log = pred.predicted_mean
pred_ci_log = pred.conf_int()

last_date = df.index[-1]
forecast_index = pd.date_range(start=last_date + pd.offsets.MonthBegin(1), periods=steps, freq='MS')

pred_mean = np.exp(pred_mean_log)
pred_ci = np.exp(pred_ci_log)

forecast_df = pd.DataFrame({
    'forecast': pred_mean.values,
    'ci_lower': pred_ci.iloc[:,0].values,
    'ci_upper': pred_ci.iloc[:,1].values
}, index=forecast_index)

plt.figure(figsize=(12,6))
plt.plot(df['Passengers'], label='Observed')
plt.plot(forecast_df['forecast'], label='Forecast')
plt.fill_between(forecast_df.index, forecast_df['ci_lower'], forecast_df['ci_upper'], alpha=0.2)
plt.title('AirPassengers: Observed and 12-month SARIMA Forecast')
plt.xlabel('Date')
plt.ylabel('Number of Passengers')
plt.legend()
plt.show()

from statsmodels.tsa.seasonal import seasonal_decompose
decomp = seasonal_decompose(df['Passengers'], model='multiplicative')
fig = decomp.plot()
fig.set_size_inches(12, 8)
plt.show()

forecast_df.to_csv("airpassengers_sarima_forecast_12mo.csv", index_label='Month')
print("Saved forecast to: airpassengers_sarima_forecast_12mo.csv")

print("\n12-month forecast:")
print(forecast_df.round(2))
```

```python
# Apply Local Outlier Factor (LOF) on any numerical dataset to detect anomalies and visualize them using matplotlib.

import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor

rng = np.random.RandomState(42)

X_inliers = 0.3 * rng.randn(200, 2)
X_inliers = np.r_[X_inliers + 2, X_inliers - 2]

X_outliers = rng.uniform(low=-6, high=6, size=(20, 2))

X = np.r_[X_inliers, X_outliers]

lof = LocalOutlierFactor(n_neighbors=20, contamination=0.1)
y_pred = lof.fit_predict(X)
scores = -lof.negative_outlier_factor_

plt.figure(figsize=(8, 6))

plt.scatter(X[y_pred == 1, 0], X[y_pred == 1, 1],
            c='blue', edgecolor='k', s=40, label="Inliers")

plt.scatter(X[y_pred == -1, 0], X[y_pred == -1, 1],
            c='red', edgecolor='k', s=60, label="Outliers")

radius = (scores.max() - scores) / (scores.max() - scores.min())
plt.scatter(X[:, 0], X[:, 1], s=1000 * radius, edgecolors='r',
            facecolors='none', alpha=0.4)

plt.title("Local Outlier Factor (LOF) - Anomaly Detection")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

```python
#  You are working as a data scientist for a power grid monitoring company. Your goal is to forecast energy demand and also detect abnormal spikes or drops in real-time consumption data collected every 15 minutes. The dataset includes features like timestamp, region, weather conditions, and energy usage.
# Explain your real-time data science workflow:
# ● How would you detect anomalies in this streaming data (Isolation Forest / LOF / DBSCAN)?
# ● Which time series model would you use for short-term forecasting (ARIMA / SARIMA / SARIMAX)?
# ● How would you validate and monitor the performance over time?
# ● How would this solution help business decisions or operations?

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from statsmodels.tsa.statespace.sarimax import SARIMAX


rng = pd.date_range("2023-01-01", periods=24*4*14, freq="15min")  # 14 days of data
np.random.seed(42)

daily_pattern = 50 + 10*np.sin(2*np.pi*rng.hour/24)

energy = daily_pattern + np.random.normal(0, 2, len(rng)) + 0.2*rng.dayofyear.values.copy()

energy = np.array(energy)

energy[200] += 50
energy[500] -= 40

df = pd.DataFrame({"timestamp": rng, "energy": energy})
df.set_index("timestamp", inplace=True)


iso = IsolationForest(contamination=0.02, random_state=42)
df["anomaly"] = iso.fit_predict(df[["energy"]])


train = df["energy"].iloc[:-48]
test = df["energy"].iloc[-48:]

model = SARIMAX(train, order=(1,1,1), seasonal_order=(1,1,1,96))
res = model.fit(disp=False)

forecast = res.get_forecast(steps=48)
forecast_mean = forecast.predicted_mean
forecast_ci = forecast.conf_int()


plt.figure(figsize=(14,6))
plt.plot(df.index, df["energy"], label="Observed", color="blue")

plt.scatter(df.index[df["anomaly"]==-1], df["energy"][df["anomaly"]==-1],
            color="red", label="Anomalies", s=40)

plt.plot(test.index, forecast_mean, label="Forecast", color="green")
plt.fill_between(test.index,
                 forecast_ci.iloc[:,0],
                 forecast_ci.iloc[:,1], color="lightgreen", alpha=0.3)

plt.title("Energy Demand: Real-time Anomaly Detection + SARIMAX Forecast")
plt.xlabel("Time")
plt.ylabel("Energy Usage (MW)")
plt.legend()
plt.show()
```
