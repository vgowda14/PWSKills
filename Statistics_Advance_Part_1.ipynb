1. What is a random variable in probability theory?
        -> In probability theory, a random variable is a numerical outcome of a random experiment. It assigns a number to each possible outcome in a sample space.

2. What are the types of random variables?
        -> Types of Random Variables:
        **Discrete Random Variable**: Takes on countable values (finite or infinite).
        **Continuous Random Variable**: Takes on uncountably infinite values within a range.

3. What is the difference between discrete and continuous distributions?
        -> The difference between discrete and continuous distributions lies in the type of values the random variable can take and how probabilities are assigned.
        **Discrete Distribution**:
        * Deals with countable outcomes (like whole numbers).
        * Each individual outcome has a non-zero probability.
        * Described using a Probability Mass Function (PMF).
        * Examples: Number of customers in a queue. and Number of heads in 10 coin tosses.

        **Continuous Distribution**:
        * Deals with uncountably infinite outcomes (like real numbers).
        * Probability at any specific point is zero; only intervals have probabilities.
        * Described using a Probability Density Function (PDF).
        Example: Height of people and Time taken to complete a task

4. What are probability distribution functions (PDF)?
        -> In probability theory, a Probability Distribution Function (PDF) describes how the probability is distributed over the values of a random variable.


5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
        -> The Difference Between PDF and CDF:
        **Probability Density Function (PDF)**:
        * Describes the likelihood of a random variable taking a specific value (for discrete) or being in an interval (for continuous).
        * Notation f(x)
        * For discrete variables: Gives P(X=x)
        * For Continuous Variables : f(x) is not a probability itself, but the area under the curve over an interval gives the probability.
        * Shape: Spiky for discrete; smooth bell or other shapes for continuous.

        **Cumulative Distribution Function (CDF)**:
        * Describes the probability that a random variable is less than or equal to a certain value.
        * Notation : F(x)
        * For Discrete Variables: Gives P(Xâ‰¤x)
        * For Continuous Variables: F(x)=âˆ« f(t)dt.
        * Shape: Always non-decreasing, S-shaped or stair-step for discrete

16. How do you calculate a Z-score, and what does it represent?
        -> A Z-score (also called a standard score) measures how many standard deviations an individual data point is from the mean of a distribution.
        Z-score represents a:
        * A Z-score of 0 means the data point is exactly at the mean.
        * A positive Z-score means the value is above the mean.
        * A negative Z-score means the value is below the mean.
        * The magnitude of the Z-score tells you how far (in standard deviations) the value is from the mean.

17. What are point estimates and interval estimates in statistics?
        -> In inferential statistics, we often want to estimate unknown population parameters (like the mean or proportion) using data from a sample. There are two main ways to do this:
        * **Point Estimate**: A point estimate is a single value that serves as a best guess for an unknown population parameter.
        * **Interval Estimate**: An interval estimate gives a range of values, derived from the sample, that is likely to contain the population parameter with a specified level of confidence.

18. What is the significance of confidence intervals in statistical analysis?
        -> A confidence interval (CI) is a range of values, derived from sample data, that is likely to contain the true population parameter with a specified level of confidence (e.g., 95%).
        **Measure of Precision**: A narrow CI implies a precise estimate of the population parameter. A wide CI suggests more variability or uncertainty.
        **Quantifies Uncertainty**: Every point estimate (like a sample mean) has sampling error. CIs acknowledge and quantify that uncertainty.
        **Helps in Decision Making**: Used to determine statistical significance. If a CI does not include the null value (like 0 for mean differences), it suggests a significant effect.
        **Better Than Point Estimates Alone**: A point estimate gives one number with no indication of how accurate it is. A CI provides a range, giving more context and confidence.
        **Used in Hypothesis Testing**: Closely linked to p-values. Example: If a 95% CI for the difference in means is (2.3, 7.5), we can reject the null hypothesis that the difference is 0 at the 0.05 level.

        Relationship Summary: A Z-score determines the number of standard deviations a data point or statistic is from the mean. A confidence interval uses the Z-score as a critical value to calculate the range in which a population parameter likely lies.

```python
# Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

def bernoulli_pdf(x, p):
    """
    Calculate the PDF (PMF) of a Bernoulli distribution.

    Parameters:
    x (int): The outcome (0 or 1)
    p (float): The probability of success (P(X=1)), where 0 <= p <= 1

    Returns:
    float: Probability of observing x
    """
    if x == 0:
        return 1 - p
    elif x == 1:
        return p
    else:
        return 0.0  # For values other than 0 or 1

# Example usage
print("P(X=0):", bernoulli_pdf(0, 0.7))
print("P(X=1):", bernoulli_pdf(1, 0.7))
```

```python
# Write a Python function to calculate Z-scores from a dataset and plot them.

import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    """
    Calculate Z-scores from a dataset and plot them.

    Parameters:
    data (list or numpy array): The dataset

    Returns:
    z_scores (numpy array): Array of Z-scores
    """
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data)

    if std == 0:
        raise ValueError("Standard deviation is zero. Cannot calculate Z-scores.")

    z_scores = (data - mean) / std

    plt.figure(figsize=(10, 5))
    plt.plot(z_scores, marker='o', linestyle='-', color='teal')
    plt.axhline(y=0, color='gray', linestyle='--')
    plt.title("Z-scores of the Dataset")
    plt.xlabel("Index")
    plt.ylabel("Z-score")
    plt.grid(True)
    plt.show()

    return z_scores

sample_data = [10, 12, 9, 15, 8, 10, 13, 7, 11, 14]
z = calculate_and_plot_z_scores(sample_data)
print("Z-scores:", z)
```

```python
# Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std):
    """
    Calculate the Z-score for a given data point.

    Parameters:
    x (float): The data point
    mean (float): Mean of the dataset
    std (float): Standard deviation of the dataset

    Returns:
    float: Z-score
    """
    if std == 0:
        raise ValueError("Standard deviation cannot be zero.")
    return (x - mean) / std

def plot_standard_normal_with_z(z_score):
    """
    Plot the standard normal distribution and mark the Z-score.

    Parameters:
    z_score (float): The Z-score to mark
    """
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, 0, 1)

    plt.figure(figsize=(10, 5))
    plt.plot(x, y, label='Standard Normal Distribution', color='blue')
    plt.fill_between(x, 0, y, where=(x < z_score), color='lightblue', alpha=0.5)
    plt.axvline(z_score, color='red', linestyle='--', label=f'Z = {z_score:.2f}')
    plt.title("Z-score Comparison to Standard Normal Distribution")
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.grid(True)
    plt.legend()
    plt.show()

data_point = 85
mean = 50
std_dev = 20

z = calculate_z_score(data_point, mean, std_dev)
print(f"Z-score for x = {data_point}: {z:.2f}")

plot_standard_normal_with_z(z)
```

```python
# Implement hypothesis testing using Z-statistics for a sample dataset.

import numpy as np
from scipy.stats import norm

population_mean = 12
sample_mean = 11.3
population_std = 1.8
n = 50
alpha = 0.05

z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(n))

p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print(f"Z-statistic: {z_stat:.2f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis: Significant difference.")
else:
    print("Fail to reject the null hypothesis: No significant difference.")
```

```python
# Create a confidence interval for a dataset using Python and interpret the result.

import numpy as np
import scipy.stats as stats

data = [22, 25, 27, 30, 24, 23, 28, 26, 29, 31]

sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
n = len(data)
confidence = 0.95

t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)
margin_of_error = t_critical * (sample_std / np.sqrt(n))

ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
# Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

import numpy as np
import scipy.stats as stats

np.random.seed(42)
mu = 100
sigma = 15
n = 50

sample_data = np.random.normal(loc=mu, scale=sigma, size=n)

sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
confidence = 0.95
alpha = 1 - confidence
df = n - 1

t_critical = stats.t.ppf(1 - alpha/2, df)
margin_of_error = t_critical * (sample_std / np.sqrt(n))

ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval for the Mean: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
# Write a Python program to compare the standard deviations of two datasets and visualize the difference.

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

data_a = np.random.normal(loc=50, scale=5, size=1000)

data_b = np.random.normal(loc=50, scale=15, size=1000)

std_a = np.std(data_a, ddof=1)
std_b = np.std(data_b, ddof=1)

print(f"Standard Deviation of Dataset A: {std_a:.2f}")
print(f"Standard Deviation of Dataset B: {std_b:.2f}")

plt.figure(figsize=(12, 5))

plt.hist(data_a, bins=30, alpha=0.6, label='Dataset A (Ïƒ â‰ˆ 5)', color='blue', edgecolor='black')
plt.hist(data_b, bins=30, alpha=0.6, label='Dataset B (Ïƒ â‰ˆ 15)', color='orange', edgecolor='black')
plt.title("Comparison of Distributions with Different Standard Deviations")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(6, 4))
plt.boxplot([data_a, data_b], labels=["Dataset A", "Dataset B"])
plt.title("Boxplot Comparison of Dataset Spread")
plt.ylabel("Values")
plt.grid(True)
plt.show()
```

```python
# Implement Z-score normalization on a dataset and visualize its transformation.

import numpy as np
import matplotlib.pyplot as plt

def z_score_normalization(data):
    """
    Apply Z-score normalization to a dataset.

    Parameters:
    data (list or np.array): Original dataset

    Returns:
    np.array: Z-score normalized data
    """
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data)

    if std == 0:
        raise ValueError("Standard deviation is zero. Cannot normalize.")

    z_scores = (data - mean) / std
    return z_scores

original_data = [55, 60, 65, 70, 75, 80, 85, 90, 95, 100]
normalized_data = z_score_normalization(original_data)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(original_data, bins=8, color='orange', edgecolor='black', alpha=0.7)
plt.title('Original Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.hist(normalized_data, bins=8, color='teal', edgecolor='black', alpha=0.7)
plt.title('Z-score Normalized Data')
plt.xlabel('Z-score')
plt.ylabel('Frequency')
plt.grid(True)

plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 4))
plt.plot(original_data, label='Original Data', marker='o')
plt.plot(normalized_data, label='Z-score Normalized', marker='s')
plt.title('Z-score Normalization Transformation')
plt.xlabel('Index')
plt.ylabel('Value / Z-score')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

import numpy as np
from scipy.stats import skew, kurtosis

def analyze_normal_distribution_skew_kurtosis(mean=0, std=1, size=1000):
    """
    Generate a dataset from a normal distribution and calculate skewness and kurtosis.

    Parameters:
    - mean: Mean of the normal distribution
    - std: Standard deviation
    - size: Number of data points

    Returns:
    - skew_val: Measure of asymmetry
    - kurt_val: Measure of tailedness (excess kurtosis)
    """
    data = np.random.normal(loc=mean, scale=std, size=size)

    skew_val = skew(data)
    kurt_val = kurtosis(data)

    print(f"Skewness: {skew_val:.4f}")
    print(f"Kurtosis (excess): {kurt_val:.4f}")

    if abs(skew_val) < 0.5:
        print(" â†’ Data is approximately symmetric.")
    elif skew_val > 0:
        print(" â†’ Data is positively skewed (tail on the right).")
    else:
        print(" â†’ Data is negatively skewed (tail on the left).")

    if abs(kurt_val) < 0.5:
        print(" â†’ Kurtosis is close to normal (mesokurtic).")
    elif kurt_val > 0:
        print(" â†’ Data is leptokurtic (heavy tails).")
    else:
        print(" â†’ Data is platykurtic (light tails).")

    return skew_val, kurt_val

analyze_normal_distribution_skew_kurtosis()

```
